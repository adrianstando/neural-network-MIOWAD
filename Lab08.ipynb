```python
df = pd.read_csv('./data/som/mnist_train.csv')

X = df.drop('label', axis=1).to_numpy()
Y = df[["label"]].to_numpy()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X, _, Y, _ = train_test_split(X, Y, train_size = 1/10, stratify = Y, random_state = 0)
```

```python
X.shape
```

### Funkcja gaussowska z siatką prostokątną

```python
mapa = SOM(width=10, height=10, vector_in_size=784, 
                   neighbourhood_function=NeighbourhoodGaussian(neighbourhood_scaler=0.8), 
                   learning_rate_function=StandardFunction(0.8), 
                   weight_initializer=RandomUniformInitializer(seed=0))
mapa.train(X, 10)
```

```python
mapa.plot_map(X, Y, legend=True)
plt.show()
```

```python
mapa.plot_map(X, Y, legend=True, type=1, init='random', learning_rate=200)
plt.show()
```

```python
import pickle

with open('som_1_mnist.obj', 'wb') as fp:
    pickle.dump(mapa, fp)
```

```python
with open('som_1_mnist.obj', 'rb') as fp:
    mapa = pickle.load(fp)
```

## Dane HAR

Pobrano z: https://archive.ics.uci.edu/ml/datasets/Human+Activity+Recognition+Using+Smartphones

```python
X_train = pd.read_csv('./data/som/UCI_HAR_X_train.txt', header=None, delim_whitespace=True)
```

```python
y_train = pd.read_csv('./data/som/UCI_HAR_y_train.txt', header=None, delim_whitespace=True)
```

```python
X_train.shape
```

```python
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
y_train = y_train - 1
```

```python
classes = ["WALKING", "WALKING UPSTAIRS", "WALKING DOWNSTAIRS", "SITTING", "STANDING", "LAYING"]
```

### Funkcja gaussowska z siatką prostokątną

```python
mapa5 = SOM(width=10, height=10, vector_in_size=561, 
                   neighbourhood_function=NeighbourhoodGaussian(neighbourhood_scaler=0.8), 
                   learning_rate_function=StandardFunction(0.8), 
                   weight_initializer=RandomUniformInitializer(seed=0))
mapa5.train(X_train, 10)
```

```python
mapa5.plot_map(X_train, y_train, Y_labels=classes, legend=True)
plt.show()
```

```python
mapa5.plot_map(X_train, y_train, Y_labels=classes, legend=True, type=1, init='random', learning_rate=200)
plt.show()
```

```python
import pickle

with open('som_5_mnist.obj', 'wb') as fp:
    pickle.dump(mapa5, fp)
```
